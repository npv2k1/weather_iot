```python
# Khai báo dữ liệu
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./dataset/weather.csv")
df = data
print (df)
```

```python
df.head()
```

```python
df_ha_noi = df[df['province'].isin(['Ha Noi'])] 
df_ha_noi
```

```python
# draw weather
df_ha_noi_weather = df_ha_noi[['min','max','date']]

# avg temp
df_ha_noi_weather['avg'] = (df_ha_noi_weather['min']+df_ha_noi_weather['max'])/2

# sort data by date
df_ha_noi_weather = df_ha_noi_weather.sort_values(by=['date'])

df_ha_noi_weather
```

```python
df_ha_noi_weather.describe()
```

```python
df_ha_noi_weather.date= pd.to_datetime(df.date)
df_ha_noi_weather2 =df_ha_noi_weather.set_index('date')
df_ha_noi_weather2
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 2, 1)
plt.title('Avg weather in Ha Noi city')
plt.plot(df_ha_noi_weather2.avg)
plt.legend(['Close price'],loc = 'upper left')
plt.subplot(1, 2, 2)
plt.title('Avg weather in Ha Noi city')
plt.bar(x=df_ha_noi_weather2.index,height=df_ha_noi_weather2['avg'])
plt.show()
```

```python
data = list(df_ha_noi_weather2["avg"])
data
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(data)
print("1. ADF : ",result[0])
print("2. P-Value : ", result[1])
print("3. Num Of Lags : ", result[2])
print("4. Num Of Observations Used For ADF Regression:", result[3])
print("5. Critical Values :")
for key, val in result[4].items():
  print("\t",key, ": ", val)
```

```python
import pmdarima as pm
from pmdarima import arima
from pmdarima import model_selection
from pmdarima import pipeline
from pmdarima import preprocessing
from pmdarima.datasets._base import load_date_example
d_value = arima.ndiffs(data,test = 'adf')
print("d value:", d_value)
```

```python
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
```

```python
from sklearn.model_selection import train_test_split

# # Tạo dữ liệu train và test với tỉ lệ 80:20
# x_train, x_test = train_test_split(data, test_size=0.2, random_state=2018)
x_train = data[:-100]
x_test = data[-100:]

# Kiểm tra từng phần tử dữ liệu mới được tạo
print("Shape of HN_weather_train:",x_train)
print("Shape of HN_weather_test:",x_test)
```

```python
stepwise_fit = auto_arima(x_train,
                          trace=True,
                          suppress_warnings=True,
                          seasonal = True, 
                          test='adf',
                          error_action='ignore',
                          m=10,
                          stepwise=True)
print(stepwise_fit.summary())
```

```python
import statsmodels.api as sm
model = sm.tsa.arima.ARIMA(x_train, order=(2,0,3))
```

```python
model = model.fit()
model.summary()
```

```python
start=len(x_train)
end=len(x_train)+len(x_test)-1
pred = model.predict(start=start,end=end)
pred
```

```python
df_guess = df_ha_noi_weather2[start:end+1]
df_guess
```

```python
s = pd.Series(pred, index =df_guess.index[0:])
s
```

```python
plt.figure(figsize=(10,6), dpi=100)
df_ha_noi_weather2['avg'][start:end+1].plot(label='Actual average temperature', legend=True)
s.plot(label='Predicted temperature', legend=True,)
```

```python
from statsmodels.graphics.tsaplots import plot_predict
plot_predict(model, start = 0, end = 101, dynamic = False);
```
